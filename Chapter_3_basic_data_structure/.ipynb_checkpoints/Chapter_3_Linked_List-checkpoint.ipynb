# Unordered List
- Unordered list is a class which points to the first Node
- a Node is either None or a class consisting a node, which stores the data, and
- a pointer to the next node

```python
# This is a node of an unordered list
class Node(object):
    def __init__(self, data):
        self.data = data
        self.next = None # this is Node or a Node
    
    # setData: obj --> void
    def setData(self, newdata):
        self.data = newdata
    
    # setNext: Node --> void
    def setNext(self, newnext):
        self.next = newnext
    
    # getData: void --> obj
    def getData(self):
        return self.data
    
    # getNext: void --> Node
    def getNext(self):
        return self.next

class UnorderedList(object):
    def __init__(self):
        self.head = None
        self.length = 0
    
    # isEmpty: void --> Bool
    def isEmpty(self):
        return self.head == None
    
    # __str__: void --> string
    def __str__(self):
        if self.head == None:
            return '[]'
        
        node = self.head
        ans = '['
        
        while node != None:
            ans += node.getData().__str__() + ', '
            node = node.getNext()
        
        return ans + '\b\b]'
            
    # add: obj --> void
    # adds item to the list at position 0
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
        self.length += 1
        
    # remove: obj --> void
    # removes item from the list
    # duplciates: lowest index item is removed
    def remove(self, item):     
        previous_node = self.head
        if previous_node != None:
            current_node = previous_node.getNext()
            
            if previous_node.getData() == item:
                self.head = current_node
                self.length -= 1
            else:
                while current_node != None:
                    if current_node.data == item:
                        previous_node.setNext(current_node.getNext())
                        self.length -= 1
                        break

                    previous_node = current_node
                    current_node = current_node.getNext()

    # search: obj --> Bool
    # True = obj in the UnorderedList
    # False = obj not in the UnorderedList
    def search(self, item):
        node = self.head
        
        while node != None:
            if node.getData() == item:
                return True
            
            node = node.getNext()
        
        return False
    
    # size: void --> int
    # returns total nodes in the UnorderedList
    def size(self):
        return self.length
    
    # indx: obj --> int
    # returns the lowest index at which the stored value is item
    def index(self, item):
        node = self.head
        
        if node == None:
            return False
        else:
            index = 0
            while node != None:
                if node.getData() == item:
                    return index
                index += 1
                node = node.getNext()
    
    # insert: int, obj --> void
    # insert item at index pos
    def insert(self, pos, item):
        node = self.head
        index = 0
        
        while index != pos:
            node = node.getNext()
        
        node.setNext(Node(item))
        node = node.getNext()
        
        self.length += 1
    
    # append: obj --> void
    # appends item to the end of the list (i.e. at index self.length)
    def append(self, item):
        node = self.head
        
        if node == None:
            self.head = Node(item)
        else:
            while node.getNext() != None:
                node = node.getNext()
            
            node.setNext(Node(item))
        
        self.length += 1

    # pop: int --> obj
    # pops the stored value at index pos
    # default pos=-1, which corresponds to the last item
    def pop(self, pos=-1):
        if pos == -1:
            return self.pop(self.length-1)
        elif pos == 0:
            ans = self.head.getData()
            self.head = self.head.getNext()
            self.length -= 1
            return ans
        else:
            index = 0
            node = self.head
        
            while index < pos-1:
                node = node.getNext()
                index += 1
            
            ans = node.getNext().getData()
            node.setNext(node.getNext().getNext())
            self.length -= 1
            return ans

            
        
```

```python
def test1():
    x = UnorderedList()
    for i in range(0,5):
        x.append(i)

    print(x)
    print(x.size())
    
    for i in range(0,5):
        x.add(i)
        
    for i in range(0,5):
        print(x.search(i))

def test2():
    x = UnorderedList()
    for i in range(0,5):
        x.add(i)
    
    print(x)
        
    for i in range(0,5):
        print(x.index(i))

def test3():
    x = UnorderedList()
    for i in range(0,5):
        x.add(i)
    
    print(x)
    
    for i in range(0,5):
        print(x.pop())
        
test1()
test2()
test3()
```
